```python
import backtrader as bt
import pandas as pd
```

```python
df = pd.read_csv('data/full/A_full.csv')
```

```python
class PandasData(bt.feeds.PandasData):
    lines = tuple([col for col in df.columns[1:]])  # Specify the column name to be used as a data line

    # Define the parameters required for the custom data feed
    params = tuple([('datetime', None)] + [(col, -1) for col in df.columns[1:]])


class MyStrategy(bt.Strategy):
    def next(self):
        # Access the close price from the data feed
        close_price = self.data.close[0]
        roe = self.data.ROE[0]

        print(f"ROE: {roe}, Close Price: {close_price}")


if __name__ == '__main__':
    # Create a sample DataFrame with OHLCV data
    df = pd.read_csv('data/full/A_full.csv')

    # Convert the 'date' column to a pandas DateTime data type
    df['Date'] = pd.to_datetime(df['Date'])

    # Set the 'date' column as the DataFrame's index
    df = df.set_index('Date')

    # Create a cerebro instance
    cerebro = bt.Cerebro()

    # Create a PandasData feed using the DataFrame
    data = PandasData(dataname=df)

    # Add the data feed to cerebro
    cerebro.adddata(data)

    # Add your strategy to cerebro
    cerebro.addstrategy(MyStrategy)

    # Run the backtest
    cerebro.run()
```
