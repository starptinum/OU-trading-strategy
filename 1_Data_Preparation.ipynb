```python
import pandas as pd
import yfinance as yf
```

```python
NDX_list = pd.read_csv('raw/nasdaq_screener_1713022569614.csv').sort_values(by='Market Cap', ignore_index=True)['Symbol'].iloc[:200]
SPX_list = pd.read_csv('raw/S&P 500 Companies (Standard and Poor 500) - basics.csv')['Symbol']
custom_list = pd.Series(['GDX', 'GLD', 'PEP', 'KO', 'SLV', 'SIL'])
stock_list = pd.concat([NDX_list, SPX_list, custom_list]).drop_duplicates().reset_index(drop=True)
```

```python
start = '2014-01-01'
end = '2024-01-01'
max_length = 0
stock_dict = {}
for stock in stock_list:
    stock_dict[stock] = yf.download(tickers=stock, start=start, end=end, interval='1d').dropna()
    if stock_dict[stock].shape[0] > max_length:
        max_length = stock_dict[stock].shape[0]
```

```python
for i, stock in enumerate(stock_list):
    if stock_dict[stock].shape[0] < max_length: # remove the stock if the stock does not have the full history
        stock_list.drop(i, inplace = True)
    else:
        stock_dict[stock].to_csv('data/stock/{}.csv'.format(stock))
stock_list.to_csv('data/stock_list.csv', index=False)
```
