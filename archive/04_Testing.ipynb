```python
import backtrader as bt
import pandas as pd

stock_list = pd.read_csv('data/selected_stock_list.csv')['Code'].iloc[:3]
df = pd.read_csv('data/full/A_full.csv')

class PandasData(bt.feeds.PandasData):
    lines = tuple([col for col in df.columns[1:]])  # Specify the column name to be used as a data line

    # Define the parameters required for the custom data feed
    params = tuple([('datetime', None)] + [(col, -1) for col in df.columns[1:]])



class MyStrategy(bt.Strategy):
    def next(self):
        # Access the close price from the data feed
        for data in self.datas:
            stock_name = data._name
            close_price = data.close[0]
            print(f"Stock: {stock_name}, Close Price: {close_price}")


if __name__ == '__main__':
    # Create sample DataFrames for two stocks
    stock_dict = {}
    for stock in stock_list:
        stock_dict[stock] = pd.read_csv('data/full/{}_full.csv'.format(stock), index_col='Date', parse_dates=['Date'])

    # Create a cerebro instance
    cerebro = bt.Cerebro()

    # Create data feeds for each stock using PandasData
    data_feed_dict = {}
    for stock in stock_list:
        data_feed_dict[stock] = PandasData(dataname=stock_dict[stock], name=stock)
        cerebro.adddata(data_feed_dict[stock])

    # Add your strategy to cerebro
    cerebro.addstrategy(MyStrategy)

    # Run the backtest
    cerebro.run()
```
